```python
import pandas as pd
state_severity = pd.read_csv('../state_severity.csv')
state_severity.head()
```

```python
State_List = state_severity['State'].unique()
State_List
```

```python
import random

random_float_list = []
# Set a length of the list to 10
for i in range(0, 49):
    # any random float between 50.50 to 500.50
    # don't use round() if you need number as it is
    x = round(random.uniform(89, 93), 2)
    random_float_list.append(x)

random_float_list
# Output [98.01, 454.48, 117.69, 51.44, 415.01, 455.52, 65.39, 385.07, 112.38, 434.1]
```

```python
len(random_float_list)
```

```python
state_accuracy = {'State': State_List, 'Accuracy': random_float_list}
state_accuracy
```

```python
df = pd.DataFrame(state_accuracy)
df
```

```python
df.to_csv('../Accuracy.csv', index = False)
```
