```python
# Import dependencies
import pandas as pd
import numpy as np
import json
import geojson
```

```python
# Read the accidents CSV into Pandas DataFrame
accidents_data = pd.read_csv('../US_Accidents_Dec20_Updated.csv')
accidents_data.head()
```

```python
accidents_data_dropped = accidents_data[['Severity', 'Start_Time', 'Start_Lat', 'Start_Lng', 'City', 'State', 'Number', 'Street', 'Description']]
accidents_data_dropped.head()
```

```python
accidents_data_dropped.rename(columns={"Start_Time": "Date/Time", "Start_Lat": "Latitude", "Start_Lng": "Longitude"}, errors="raise", inplace=True)
accidents_data_dropped.head()
```

```python
arkansas_df = accidents_data_dropped.loc[accidents_data_dropped['State'] == "AR"]
arkansas_df.head()
```

```python
arkansas_sample_df = arkansas_df.sample(2000)
arkansas_sample_df.head()
```

```python
arkansas_sample_df = arkansas_sample_df.round(decimals=4)
arkansas_sample_df.head()
```

```python
arkansas_sample_df.info()
```

```python
arkansas_sample_df['Number'] = arkansas_sample_df['Number'].fillna(0)
```

```python
arkansas_sample_df['Number'] = arkansas_sample_df['Number'].map("{:.0f}".format)
```

```python
arkansas_sample_df['Number'].replace({"0": ""}, inplace=True)
```

```python
arkansas_sample_df.reset_index(drop=True, inplace=True)
arkansas_sample_df.head()
```

```python
arkansas_sample_df.to_csv('../State_sample_csv/Arkansas_sample.csv', encoding='UTF-8', index=False)
```
