```python
# Import dependencies
import pandas as pd
import numpy as np
import json
import geojson
```

```python
# Read the accidents CSV into Pandas DataFrame
accidents_data = pd.read_csv('../US_Accidents_Dec20_Updated.csv')
accidents_data.head()
```

```python
accidents_data_dropped = accidents_data[['Severity', 'Start_Time', 'Start_Lat', 'Start_Lng', 'City', 'State', 'Number', 'Street', 'Description']]
accidents_data_dropped.head()
```

```python
accidents_data_dropped.rename(columns={"Start_Time": "Date/Time", "Start_Lat": "Latitude", "Start_Lng": "Longitude"}, errors="raise", inplace=True)
accidents_data_dropped.head()
```

```python
california_df = accidents_data_dropped.loc[accidents_data_dropped['State'] == "CA"]
california_df.head()
```

```python
california_sample_df = california_df.sample(2000)
california_sample_df.head()
```

```python
california_sample_df = california_sample_df.round(decimals=4)
california_sample_df.head()
```

```python
california_sample_df.info()
```

```python
california_sample_df['Number'] = california_sample_df['Number'].fillna(0)
```

```python
california_sample_df['Number'] = california_sample_df['Number'].map("{:.0f}".format)
```

```python
california_sample_df['Number'].replace({"0": ""}, inplace=True)
```

```python
california_sample_df.reset_index(drop=True, inplace=True)
california_sample_df.head()
```

```python
california_sample_df.to_csv('../State_sample_csv/California_sample.csv', encoding='UTF-8', index=False)
```
