```python
import pandas as pd
us_accidents_df = pd.read_csv("../US_Accidents_Dec20_Updated.csv")
us_accidents_df.head()
```

```python
us_df = us_accidents_df[['State', 'Severity', 'Start_Time']]
us_df
```

```python
us_df[['Year', 'Month', 'Day_Time']] = us_df.Start_Time.str.split('-', expand=True,)
us_df

# season_one_df[['City', 'State']] = season_one_df.Hometown.str.split(",",expand=True,)
```

```python
State_List = us_df['State'].unique()
State_List
```

```python
us_df[['Day', 'Time']] = us_df.Day_Time.str.split(' ', expand=True,)
us_df
```

```python
us_df = us_df[['State', 'Severity', 'Year', 'Month', 'Day', 'Time']]
us_df
```

```python
us_df['Month'] = us_df['Month'].replace(to_replace=['01', '02', '03', '04', '05', '06', '07', '08', '09', '10', '11', '12'],
                                        value=["January", "February", "March", "April", "May", "June", "July", "August",
                                               "September", "October", "November", "December"])
us_df
```

```python
us_df['Counter'] =1       #initially, set that counter to 1.
group_data = us_df.groupby(['State', 'Severity', 'Year', 'Month', 'Day'])['Counter'].sum() #sum function
state_severity_df = pd.DataFrame(group_data)
state_severity_df
```

```python
state_severity_df.to_csv("../state_severity.csv")
```
