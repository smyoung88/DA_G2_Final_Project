```python
import pandas as pd
data = pd.read_csv('../US_Accidents_Dec20_Updated.csv')
accidents_data = data[['Severity', 'Start_Time', 'Start_Lat', 'Start_Lng', 'City', 'State', 'Number', 'Street', 'Description']]
accidents_data.rename(columns={"Start_Time": "Date/Time", "Start_Lat": "Latitude", "Start_Lng": "Longitude"}, errors="raise", inplace=True)
accidents_data.head()
```

```python
test_data = accidents_data[0:1000]
test_data
```

```python
states = set(accidents_data['State'])
states = sorted(list(states))
```

```python
d = {}
for state in states:
    d[state] = pd.DataFrame(columns = ['Severity', 'Date/Time', 'Latitude', 'Longitude', 'City', 'State', 'Number', 'Street', 'Description'])
```
