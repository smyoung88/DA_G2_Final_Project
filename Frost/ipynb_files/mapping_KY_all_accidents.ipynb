```python
# Import dependencies
import pandas as pd
```

```python
# Read the accidents CSV into Pandas DataFrame
accidents_data = pd.read_csv('../US_Accidents_Dec20_Updated.csv')
accidents_data.head()
```

```python
accidents_data_dropped = accidents_data[['Severity', 'Start_Time', 'Start_Lat', 'Start_Lng', 'City', 'State', 'Number', 'Street', 'Description']]
accidents_data_dropped.head()
```

```python
accidents_data_dropped.rename(columns={"Start_Time": "Date/Time", "Start_Lat": "Latitude", "Start_Lng": "Longitude"}, errors="raise", inplace=True)
accidents_data_dropped.head()
```

```python
kentucky_df = accidents_data_dropped.loc[accidents_data_dropped['State'] == "KY"]
kentucky_df
```

```python
kentucky_df = kentucky_df.round(decimals=4)
kentucky_df.head()
```

```python
kentucky_df.info()
```

```python
kentucky_df['Number'] = kentucky_df['Number'].fillna(0)
```

```python
kentucky_df['Number'] = kentucky_df['Number'].map("{:.0f}".format)
```

```python
kentucky_df['Number'].replace({"0": ""}, inplace=True)
```

```python
kentucky_df.to_csv('../State_csv_files/Kentucky.csv', encoding='UTF-8', index=False)
```
